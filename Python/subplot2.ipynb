```python
import matplotlib.pyplot as plt
x = [1,2,3]
y = [4,5,6]
z = [7,8,9]
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.plot3D(x,y,z)
plt.rcParams["grid.color"] = 'blue'
plt.show()
```
