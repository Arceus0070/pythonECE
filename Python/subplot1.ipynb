```python
import matplotlib.pyplot as plt

def plot_income(names, incomes):
    plt.bar(names, incomes)
    plt.xlabel('Names')
    plt.ylabel('Income')
    plt.title('Income Comparison')
    plt.show()


names = []
incomes = []

for i in range(3):
    name = input("Enter name: ")
    income = float(input("Enter income: "))
    names.append(name)
    incomes.append(income)

plot_income(names, incomes)
```

```python
import matplotlib.pyplot as plt

def plot_sales(products, sales):
    plt.pie(sales, labels=products, autopct='%1.1f%%')
    plt.title('Product Sales Comparison')
    plt.axis('equal')
    plt.show()

products = []
sales = []

for i in range(3):
    product = input("Enter product name: ")
    sale = float(input("Enter sales percentage: "))
    products.append(product)
    sales.append(sale)

plot_sales(products, sales)
```

```python
import matplotlib.pyplot as plt

def plot_bar_graph(students, marks):
    colors = ['red', 'blue', 'green', 'yellow', 'orange']
    
    plt.bar(students, marks, color=colors)
    plt.xlabel('Students')
    plt.ylabel('Marks')
    plt.title('Student Marks')
    plt.show()

if __name__ == '__main__':
    students = ['Alice', 'Bob', 'Charlie', 'David', 'Eve']
    marks = []
    
    for student in students:
        mark = int(input(f"Enter the marks for {student}: "))
        marks.append(mark)
    
    plot_bar_graph(students, marks)
```
